# Doubly linked list


```python
class Node:
    def __init__(self,value):
        self.value=value
        self.next=None
        self.prev=None
        
        
class LinkedLlist:
    def __init__(self):
        self.head=None
        self.tail=None
        self.length=0
        
    def creation(self,value):
        new_node=Node(value)
        self.head=new_node
        self.tail=new_node
        self.length+=1
        
    def __iter__(self):
        node=self.head
        while node is not None:
            yield node
            node=node.next
            
        
    def addBegin(self,value):
        if self.head is None:
            new_node=Node(value)
            self.head=new_node
            self.tail=new_node
        else:
            new_node=Node(value)
            self.head.prev=new_node
            new_node.next=self.head
            self.head=new_node  
            
    def addEnd(self,value):
        if self.head is None:
            new_node=Node(value)
            self.head=new_node
            self.tail=new_node
        else:
            new_node=Node(value)
            self.tail.next=new_node
            new_node.prev=self.tail
            new_node=self.tail
            
    def addMiddle(self,value,index):
        new_node=Node(value)
        temp=self.head
        ind=0
        while ind<index-1:
            temp=temp.next
            ind+=1
        new_node.next=temp.next
        new_node.prev=temp
        new_node.next.prev=new_node
        temp.next=new_node
        
    def  traverse(self):
        if self.head is None:
            print('linked list is empty')
        else:
            temp=self.head
            while temp is not None:
                print(temp.value)
                temp=temp.next
                
    def search(self,value):
        if self.head is None:
            print('Linked list empty')
        else:
            temp=self.head
            index=0
            while temp is not None:
                if temp.value==value:
                    return index
                index=index+1
                temp=temp.next
                    
    def traverseReverse(self):
        if self.head is None:
            print('linked list empty')
        else:
            temp=self.tail
            while temp:
                print(temp.value)
                temp=temp.prev


new=LinkedLlist()

new.addBegin(20)
new.addBegin(30)
new.addBegin(10)
new.addMiddle(500,2)
new.addEnd(70)
print(new.length)
new.traverse()
print(new.search(30))
print([i.value for i in new])
new.traverseReverse()
```
