```python
#Define a function with takes two dictionaries as parameters and merge them and sum the values of common keys.
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 3, 'c': 4, 'd': 5}
l1=list(dict1.items())
l2=list(dict2.items())
# print(l1)
# print(l2)
l3={}
for i in l1:
    for j in l2:
        i1=list(i)
        j1=list(j)
        if i1[0]==j1[0]:
            s=i1[1]+j1[1]
            l3[i1[0]]=s
            
combined_dict = {**dict1, **dict2, **l3}
print(combined_dict)
            
            
```

```python
#or
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 3, 'c': 4, 'd': 5}
d3={}
for i in dict1:
    if i in dict2:
        d3[i]=dict1[i]+dict2[i]
    else:
        d3[i]=dict1[i]
        
for i in dict2:
    if i not in dict1:
        d3[i]=dict2[i]
print(d3)
    
```

```python
#or
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 3, 'c': 4, 'd': 5}

result = dict1.copy()
for key, value in dict2.items():
    result[key] = result.get(key, 0) + value
print(result)
```
