```python
class Node:
    def __init__(self,value):
        self.value=value
        self.next=None
        
class CircularLinkedList:
    def __init__(self):
        self.head=None
        self.tail=None
        self.length=0
        
    def __iter__(self):
        node=self.head
        while node is not None:
            yield node
            node=node.next
            if node==self.head:
                break
          
            
    def creation(self,value):
        new_node=Node(value)
        new_node.next=new_node
        self.head=new_node
        self.tail=new_node
        
        
    def addAtbegin(self,value):
        if self.head is None:
            new_node=Node(value)
            new_node.next=new_node
            self.head=new_node
            self.tail=new_node
        else:
            new_node=Node(value)
            new_node.next=self.head
            self.head=new_node
            self.tail.next=new_node
            self.length+=1
            
            
    def addAtend(self,value):
        if self.head is None:
            new_node=Node(value)
            new_node.next=new_node
            self.head=new_node
            self.tail=new_node
        else:
            new_node=Node(value)
            new_node.next=self.tail.next
            self.tail.next=new_node
            self.tail=new_node
            self.length+=1
            
            
    def addAtmiddle(self,value,index):
        if self.head is None:
            new_node=Node(value)
            new_node.next=new_node
            self.head=new_node
            self.tail=new_node
        else:
            if index==0:
                self.addAtbegin(value)
            else:
                new_node=Node(value)
                temp=self.head
                for i in range(index-1):
                    temp=temp.next
                new_node.next=temp.next
                temp.next=new_node
                
                
                
    def traversal(self):
        if self.head is None:
            print('linked list empty')
        else:
            temp=self.head
            while temp:
                print(temp.value)
                temp=temp.next
                if temp.next==self.head:
                    break
                    
    def search(self,value):
        temp=self.head
        index=0
        while temp:
            if temp.value==value:
                return index
            else:
                temp=temp.next
                index=index+1
                if temp.next==self.head:
                    return 'node not exist'

                
    def deleteAtbegin(self):
        if self.head==self.tail:
            self.head.next=None
            self.head=None
            self.tail=None
        else:
            self.head=self.head.next
            self.tail.next=self.head
         
    def deleteAtEnd(self):
        if self.head==self.tail:
            self.head.next=None
            self.head=None
            self.tail=None
        else:
            self.head=self.head.next
            self.tail.next=self.head
            
            
    def delteEntire(self):
        self.head=None
        self.tail=None
       #
         
    def deleteAtMiddle(self, index):
        if self.head is None:
            print('Linked list is empty')
            return

        if index == 0:
            self.deleteAtBegin()
        else:
            current = self.head
            prev = None
            count = 0

            while True:
                prev = current
                current = current.next
                count += 1

                if count == index:
                    prev.next = current.next

                    if current == self.tail:
                        self.tail = prev

                    break

                if current == self.head:
                    print('Index out of range')
                    return
 





new1=CircularLinkedList()
new1.creation(100)
new1.addAtbegin(200)
new1.addAtend(300)
new1.addAtend(300)
new1.addAtend(300)
new1.addAtmiddle(500,1)
new1.deleteAtMiddle(1)
new1.deleteAtbegin()
new1.deleteAtEnd()
new1.traversal()
print(new1.search(500))
new1.delteEntire()
print([i.value for i in new1])
```
