# Reversed linkedlist

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0
    
    def __str__(self):
        temp_node = self.head
        result = ''
        while temp_node is not None:
            result += str(temp_node.value)
            if temp_node.next is not None:
                result += ' -> '
            temp_node = temp_node.next
        return result
    
    def append(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length += 1
        
    def deleteatend(self):
        temp=self.head
        while temp.next is not self.tail:
            temp=temp.next
        poped=temp.next
        temp=self.tail
        temp.next=None
        return poped
        self.length=self.length-1
        
    def reverse(self):
        prev=None
        curr=self.head
        while curr is not None:
            forward=curr.next
            curr.next=prev
            prev=curr
            curr=forward

        self.head = prev
        self.tail = self.head  # Update the tail to the new head node
        return self.head

        
        
new_link = LinkedList()
new_link.append(10)
new_link.append(20)
new_link.append(39)

print("Original Linked List:", new_link)
new_link.reverse()
print("Reversed Linked List:", new_link)
```
