# tuple in memory




```python
#Access elements of tuple
t3=(12,23,4,5,6,7,8)
print(t3[3])
print(t3[-1])#last element in tuple
```

```python
#using slice operator
print(t3[:])
print(t3[1:])
print(t3[2:3])
print(t3[:4])
```
