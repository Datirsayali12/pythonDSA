```python
#buuble sort


def bubblesort(list1):
    for i in range(len(list1)-1):
        for j in range (len(list1)-i-1):
            if list1[j]>list1[j+1]:
                list1[j],list1[j+1]=list1[j+1],list1[j]
    print(list1)
    
list1=[6,2,1,4,0,1]
bubblesort(list1)         

                
```

# Insertion Sort:

```python
list1=[1,0,3,2]

for i in range(1,len(list1)):
    curr=list1[i]
    pos=i
    while list1[pos]<list1[pos-1] and pos>0:
        list1[pos]=list1[pos-1]
        pos=pos-1
    list1[pos]=curr
    
print(list1)
        
```

# Selecting First Number as Pivot:

```python
#ASCENDING ORDER:
def pivot_place(list1,first,last):
    pivot=list1[first]
    left=first+1
    right=last
    
    while True:
        while left<=right and list1[left]<=pivot:
            left=left+1
        while left<=right and list1[right]>=pivot:
            right=right-1

        if left>right:
            break
        else:
            list1[right],list1[left]=list1[left],list1[right]
    list1[first],list1[right]=list1[right],list1[first]
    return right

def quicksort(list1,first,last):
    if first<last:
        p=pivot_place(list1,first,last)
        quicksort(list1,first,p-1)
        quicksort(list1,p+1,last)
        
list1=[0,3,1,2,4,5]
quicksort(list1,0,len(list1)-1)

print(list1)       
    
```

# median of three values:(pivot)

```python
import statistics


def pivot_place(list1,first,last):
    low=list1[first]
    high=list1[last]
    mid=(first+last)//2
    pivot_val=statistics.median([low,list1[mid],high])
    if pivot_val==low:
        pindex=first
    elif pivot_val==high:
        pindex=last
    else:
        pindex=mid
        
    list1[last],list1[pindex]=list1[pindex],list1[last]    
    pivot=list1[last]
    right=last-1
    left=first
    while True:
        while left<=right and list1[left]<=pivot:
            left=left+1
        while left<=right and list1[right]>=pivot:
            right=right-1

        if left>right:
            break
        else:
            list1[right],list1[left]=list1[left],list1[right]
    list1[last],list1[left]=list1[left],list1[last]
    return left

def quicksort(list1,first,last):
    if first<last:
        p=pivot_place(list1,first,last)
        quicksort(list1,first,p-1)
        quicksort(list1,p+1,last)
        
list1=[0,3,1,2,4,5]
quicksort(list1,0,len(list1)-1)

print(list1)       
    
```

```python
list1=[1,2,3,4,5,10,12]
list1.sort()
low=0
high=len(list1)-1

n=int(input('Enter element'))
found=False
while low<=high and not found:
    mid=(low+high)//2
    if list1[mid]==n:
        found=True
    elif n>list1[mid]:
        low=mid+1
        
    elif n<list1[mid]:
        high=mid-1
        
if found==True:
    print('key found')
else:
    print('not found')
        
```

```python
#Using Recursion

def binarysearch(list1,low,high,ele):
    if low<=high:
        mid=(low+high)//2

        if list1[mid]==ele:
            return mid
        elif ele>list1[mid]:
            return binarysearch(list1,mid+1,high,ele)
        else:
            return binarysearch(list1,low,mid-1,ele)
    return -1

list1=[1,2,3,4,5]
low=0
high=len(list1)
ele=2
result=binarysearch(list1,low,high,ele)

if result==-1:
    print('not found')
else:
    print('found')
```
